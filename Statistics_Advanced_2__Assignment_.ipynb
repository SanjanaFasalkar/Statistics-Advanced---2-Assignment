-  Question 1:
-  What is hypothesis testing in statistics?
-  answers:-Hypothesis Testing in Statistics

-  Hypothesis testing is a statistical method used to make decisions or draw conclusions about a population based on sample data.

-  It helps us answer questions like:

-  ‚ÄúIs this new drug more effective than the old one?‚Äù

-  ‚ÄúDo men and women spend the same amount of time on social media?‚Äù

-  ‚ÄúIs the average salary in 2025 higher than in 2020?‚Äù

-  Key Concepts in Hypothesis Testing

-  Hypotheses

-  Null Hypothesis (H‚ÇÄ): The default assumption (no effect, no difference).
Example: The new drug has the same effect as the old drug.

-  Alternative Hypothesis (H‚ÇÅ or Ha): What you want to prove (an effect, a difference).
Example: The new drug is more effective than the old drug.

-  Significance Level (Œ±):

-  The probability of rejecting the null hypothesis when it is actually true (Type I error).

Commonly used: Œ± = 0.05 (5% risk).

Test Statistic:

-  A value calculated from the sample data (e.g., z-score, t-score).

Measures how far the sample result is from what we‚Äôd expect under H‚ÇÄ.

P-value:

-  The probability of observing results as extreme as (or more extreme than) the sample result, assuming H‚ÇÄ is true.

If p ‚â§ Œ±, reject H‚ÇÄ ‚Üí evidence supports H‚ÇÅ.

If p > Œ±, fail to reject H‚ÇÄ ‚Üí insufficient evidence to support H‚ÇÅ.

Decision:

-  Based on the test statistic or p-value, decide whether to reject or fail to reject the null hypothesis.

Example

-  Suppose a company claims that the average weight of their cereal boxes is 500 g. A consumer group suspects it‚Äôs less.

H‚ÇÄ: Œº = 500 g (average weight is 500 g).

H‚ÇÅ: Œº < 500 g (average weight is less than 500 g).

Collect a sample, calculate the mean, run a t-test, and check the p-value.

-  If p < 0.05 ‚Üí Reject H‚ÇÄ ‚Üí Evidence that cereal boxes weigh less than 500 g.


-  Question 2:
-   What is the null hypothesis, and how does it differ from the alternative
hypothesis?
-answers:- Null Hypothesis (H‚ÇÄ):

-  The null hypothesis is a default assumption in statistics that there is no effect, no difference, or no relationship in the population.

-  It acts as the ‚Äústatus quo‚Äù or baseline claim.

-  Example: If you‚Äôre testing a new medicine, the null hypothesis might state:
‚ÄúThe medicine has no effect on patient recovery compared to a placebo.‚Äù

-  Alternative Hypothesis (H‚ÇÅ or Ha):

-  The alternative hypothesis is what you want to prove. It suggests that there is an effect, difference, or relationship in the population.

-  Example: In the same medicine study, the alternative hypothesis would be:
‚ÄúThe medicine improves patient recovery compared to a placebo.‚Äù

-  Key Differences:
-  Aspect	Null Hypothesis (H‚ÇÄ)	Alternative Hypothesis (H‚ÇÅ)
Meaning	No effect, no difference, status quo	Some effect, difference, or relationship
Role	Default assumption to be tested	What the researcher aims to support
Symbol	H‚ÇÄ	H‚ÇÅ or Ha
Example (coin toss)	The coin is fair (p = 0.5)	The coin is biased (p ‚â† 0.5)

-  In hypothesis testing, we never "prove" the null hypothesis; we either reject it (in favor of H‚ÇÅ) or fail to reject it (keep H‚ÇÄ).



-  Question 3:
-  Explain the significance level in hypothesis testing and its role in deciding
the outcome of a test.
-  answers:- The significance level (denoted by Œ±) in hypothesis testing is the threshold probability used to decide whether to reject the null hypothesis (H‚ÇÄ).

-  Key Points:

-  Definition:

-  The significance level (Œ±) is the probability of making a Type I error‚Äîrejecting the null hypothesis when it is actually true.

-  Common choices are 0.05 (5%), 0.01 (1%), or 0.10 (10%).

-  Role in Testing:

-  After performing a statistical test, we calculate a p-value, which represents the probability of obtaining the observed data (or more extreme results) if the null hypothesis is true.

-  Decision rule:

-  If p-value ‚â§ Œ± ‚Üí Reject H‚ÇÄ (evidence supports the alternative hypothesis).

-  If p-value > Œ± ‚Üí Fail to reject H‚ÇÄ (not enough evidence against H‚ÇÄ).

-  Example:

Suppose Œ± = 0.05. This means we are willing to accept a 5% risk of wrongly rejecting H‚ÇÄ.

If the p-value = 0.03, since 0.03 < 0.05, we reject H‚ÇÄ.

-  Interpretation:

-  A smaller Œ± reduces the chance of a Type I error (false positive) but increases the chance of a Type II error (failing to reject a false H‚ÇÄ).

-  The choice of Œ± depends on the context:

-  Medical trials: Often use Œ± = 0.01 (very strict, to avoid false claims).

-  Exploratory research: May use Œ± = 0.10 (more lenient).


-  Question 4:
- What are Type I and Type II errors? Give examples of each.
- answers:-ype I and Type II Errors in Hypothesis Testing

-  When we perform hypothesis testing, we are testing the null hypothesis (H‚ÇÄ) against the alternative hypothesis (H‚ÇÅ).
But since we rely on sample data, there‚Äôs always a chance of making mistakes:

-  Type I Error (False Positive)

-  Definition: Rejecting the null hypothesis (H‚ÇÄ) when it is actually true.

-  Meaning: You conclude there is an effect/difference when in reality there isn‚Äôt.

-  Probability of Type I Error: Denoted by Œ± (alpha), usually the significance level (e.g., 0.05).

-  Example:

-  A medical test for a disease says a person is sick (rejects H‚ÇÄ: "the person is healthy"), but in reality, the person is healthy.

-  In court, convicting an innocent person (rejecting H‚ÇÄ: "the defendant is innocent") is a Type I error.

-  Type II Error (False Negative)

-  Definition: Failing to reject the null hypothesis (H‚ÇÄ) when the alternative hypothesis (H‚ÇÅ) is actually true.

-  Meaning: You conclude there is no effect/difference when in reality there is.

-  Probability of Type II Error: Denoted by Œ≤ (beta). Power of the test is 1 ‚àí Œ≤.

-  Example:

-  A medical test fails to detect a disease (fails to reject H‚ÇÄ: "the person is healthy") when the person is actually sick.

-  In court, letting a guilty person go free (failing to reject H‚ÇÄ: "the defendant is innocent") is a Type II error.

-  Question 7:
-  Implement hypothesis testing using Z-statistics for a sample dataset in Python. Show the Python code and interpret the results. sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6, 50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5, 50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9, 50.3, 50.4, 50.0, 49.7, 50.5, 49.9]
-  answers:- We‚Äôll perform a one-sample Z-test using the given dataset.

-  Problem Setup:

-  Null Hypothesis (H‚ÇÄ): The population mean = 50

-  Alternative Hypothesis (H‚ÇÅ): The population mean ‚â† 50 (two-tailed test)

-  Significance Level (Œ±): 0.05 (commonly used)

-  Since we are using Z-statistics, we assume the population standard deviation is known.
If it is unknown, we would use a T-test instead.

-  For demonstration, let‚Äôs assume œÉ = 1 (population std. dev.).


        import numpy as np
        from scipy.stats import norm

        # Sample data
        sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

      # Step 1: Calculate sample statistics
      sample_mean = np.mean(sample_data)
      n = len(sample_data)
      sigma = 1  # assumed population standard deviation
      mu_0 = 50  # hypothesized population mean

      # Step 2: Compute Z-statistic
      z = (sample_mean - mu_0) / (sigma / np.sqrt(n))

      # Step 3: Calculate p-value (two-tailed test)
      p_value = 2 * (1 - norm.cdf(abs(z)))

      # Step 4: Print results
      print("Sample Mean:", round(sample_mean, 3))
      print("Z-statistic:", round(z, 3))
      print("p-value:", round(p_value, 4))

      # Step 5: Decision
      alpha = 0.05
      if p_value < alpha:
      print("Reject the Null Hypothesis (H0).")
      else:
      print("Fail to Reject the Null Hypothesis (H0).")

-   Interpretation of Results:

-  The Z-statistic tells us how many standard errors away the sample mean is -  from the hypothesized mean.

-  The p-value shows the probability of observing such a result if the null hypothesis is true.

-  If p-value < 0.05, we reject H‚ÇÄ (evidence the mean ‚â† 50).

-  If p-value ‚â• 0.05, we fail to reject H‚ÇÄ (no strong evidence against mean = 50).


-  Question 8:
-   Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.
- answers:- Here‚Äôs a complete Python script to simulate data from a normal distribution, compute the 95% confidence interval (CI) for the mean, and plot the data using Matplotlib:

         import numpy as np
         import matplotlib.pyplot as plt
         import scipy.stats as stats

         # Step 1: Simulate data from a normal distribution
         np.random.seed(42)  # for reproducibility
         data = np.random.normal(loc=50, scale=5, size=1000)  # mean=50, std=5, sample size=1000

         # Step 2: Calculate sample mean and standard error
         mean = np.mean(data)
         sem = stats.sem(data)  # standard error of mean

         # Step 3: Compute 95% confidence interval
         confidence_level = 0.95
         ci = stats.norm.interval(confidence_level, loc=mean, scale=sem)

         print(f"Sample Mean: {mean:.2f}")
         print(f"95% Confidence Interval for Mean: {ci}")

         # Step 4: Plot histogram of the data
         plt.figure(figsize=(8, 5))
         plt.hist(data, bins=30, density=True, alpha=0.6, color="skyblue", edgecolor="black")
         plt.axvline(mean, color="red", linestyle="--", label=f"Mean = {mean:.2f}")
         plt.axvline(ci[0], color="green", linestyle="--", label=f"95% CI Lower = {ci[0]:.2f}")
         plt.axvline(ci[1], color="green", linestyle="--", label=f"95% CI Upper = {ci[1]:.2f}")

         plt.title("Normal Distribution with 95% Confidence Interval")
         plt.xlabel("Value")
         plt.ylabel("Density")
         plt.legend()
         plt.show()

-   Explanation:

-  np.random.normal generates random samples from a normal distribution.

-  Mean (loc) = 50

-  Standard deviation (scale) = 5

-   Sample size = 1000

-  np.mean(data) gives the sample mean.

-  stats.sem(data) calculates the standard error of the mean.

-  stats.norm.interval computes the 95% confidence interval for the population mean.

-  The plot shows the histogram of simulated data, the mean (red dashed line), and confidence interval bounds (green dashed lines).
